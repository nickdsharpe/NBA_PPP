```python
import pandas as pd
import time
import json
import numpy as np
import nbimporter
from PPP import PPP
np.seterr(invalid='ignore')
newGame = 'Heat2.27.23'
```

# Read in new files and load overall files

```python
try:
    newEmbiidFile = (f'Embiid/Embiid{newGame}.csv')
    newEmbiid = pd.read_csv(newEmbiidFile)
    newEmbiid = newEmbiid.dropna()
    newEmbiid = newEmbiid.set_index('Shot Type')
    newEmbiid = newEmbiid.astype('int')
except(FileNotFoundError):
    print('No Embiid File Found.')

embiidFile = ('!Data/Embiid.csv')
embiid = pd.read_csv(embiidFile)
embiid = embiid.dropna()
embiid = embiid.set_index('Shot Type')
embiid = embiid.astype('int')
```

```python
try:
    newHardenFile = (f'Harden/Harden{newGame}.csv')
    newHarden = pd.read_csv(newHardenFile)
    newHarden = newHarden.dropna()
    newHarden = newHarden.set_index("Shot Type")
    newHarden = newHarden.astype('int')
    newHarden
except(FileNotFoundError):
    print('No Harden file found')

hardenfile = ('!Data/Harden.csv')
harden = pd.read_csv(hardenfile)
harden = harden.dropna()
harden = harden.set_index('Shot Type')
harden = harden.astype('int')
```

```python
try:
    newHarrisFile = (f'Harris/Harris{newGame}.csv')
    newHarris = pd.read_csv(newHarrisFile)
    newHarris = newHarris.dropna()
    newHarris = newHarris.set_index("Shot Type")
    newHarris = newHarris.astype('int')
    newHarris
except(FileNotFoundError):
    print('No Harris file found')

harrisfile = ('!Data/Harris.csv')
harris = pd.read_csv(harrisfile)
harris = harris.dropna()
harris = harris.set_index('Shot Type')
harris = harris.astype('int')
```

```python
try:
    newMaxeyFile = (f'Maxey/Maxey{newGame}.csv')
    newMaxey = pd.read_csv(newMaxeyFile)
    newMaxey = newMaxey.dropna()
    newMaxey = newMaxey.set_index("Shot Type")
    newMaxey = newMaxey.astype('int')
    newMaxey
except(FileNotFoundError):
    print('No Maxey file found')

maxeyfile = ('!Data/Maxey.csv')
maxey = pd.read_csv(maxeyfile)
maxey = maxey.dropna()
maxey = maxey.set_index('Shot Type')
maxey = maxey.astype('int')
```

```python
if harden.columns.equals(newHarden.columns):
    harden = harden.add(newHarden)
    harden.to_csv('!Data/Harden.csv')
else:
    print("Columns do not match")
```

```python
if harris.columns.equals(newHarris.columns):
    harris = harris.add(newHarris)
    harris.to_csv('!Data/Harris.csv')
else:
    print("Columns do not match")
```

```python
if maxey.columns.equals(newMaxey.columns):
    maxey = maxey.add(newMaxey)
    maxey.to_csv('!Data/Maxey.csv')
else:
    print("Columns do not match")
```
