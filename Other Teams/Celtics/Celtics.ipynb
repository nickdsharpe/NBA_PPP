# Read in new files and load overall files

```python
try:
    newBrogdonFile = (f'Brogdon/Brogdon{newGame}.csv')
    newBrogdon = pd.read_csv(newBrogdonFile)
    newBrogdon = newBrogdon.dropna()
    newBrogdon = newBrogdon.set_index("Shot Type")
    newBrogdon = newBrogdon.astype('int')
    newBrogdon
except(FileNotFoundError):
    print('No Brogdon file found')

brogdonfile = ('!Data/Brogdon.csv')
brogdon = pd.read_csv(brogdonfile)
brogdon = brogdon.dropna()
brogdon = brogdon.set_index('Shot Type')
brogdon = brogdon.astype('int')
```
