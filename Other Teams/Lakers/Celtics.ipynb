```python
import pandas as pd
import time
import numpy as np
np.seterr(invalid='ignore')
import nbimporter
from PPP import PPP

newGame = 'Game2Nuggets'
```

```python
players = ['davis', 'lebron', 'reaves', 'rui', 'russell', 'schroder', 'vando', 'walker', 'opp', 'team']
```

```python
def load_data(players, add=False):
    
    new_player_data = {}
    ovr_player_data = {}
    series_player_data = {}
    
    for player in players:
        
        try:
            newplayerFile = (f'{player.title()}/{newGame + player.title()}.csv')
            newplayer = pd.read_csv(newplayerFile)
            newplayer = newplayer.dropna()
            newplayer = newplayer.set_index('Shot Type')
            newplayer = newplayer.astype('int')
            new_player_data[player] = newplayer
            
        except(FileNotFoundError):
            print(f'No {player.title()} file found.')

        # Load overall data
        ovrplayerFile = (f'!Data/{player.title()}.csv')
        ovrplayer = pd.read_csv(ovrplayerFile)
        ovrplayer = ovrplayer.dropna()
        ovrplayer = ovrplayer.set_index('Shot Type')
        ovrplayer = ovrplayer.astype('int')
        ovr_player_data[player] = ovrplayer
        
        
        if add == True:

            if newplayer.columns.equals(ovrplayer.columns):
                ovrplayer = ovrplayer.add(newplayer)
                ovr_player_data[player] = ovrplayer
                ovr_player_data[player].to_csv(f'!Data/{player.title()}.csv')
            else:
                print(f"{player.title()}'s columns do not match")
            
    
    return new_player_data, ovr_player_data   

new_player_data, ovr_player_data  = load_data(players, add=False)
```

```python
newTeam = new_player_data['davis']
newTeam = newTeam.add(new_player_data['lebron'])
newTeam = newTeam.add(new_player_data['reaves'])
newTeam = newTeam.add(new_player_data['rui'])
newTeam = newTeam.add(new_player_data['russell'])
newTeam = newTeam.add(new_player_data['schroder'])
newTeam = newTeam.add(new_player_data['vando'])
newTeam = newTeam.add(new_player_data['walker'])
newTeam.to_csv(f'../../Nuggets/Opponent/Game2Lakers.csv')
newTeam.to_csv(f'Team/{newGame}Team.csv')
```

```python
if ovr_player_data['team'].columns.equals(newTeam.columns):
    ovr_player_data['team'] = ovr_player_data['team'].add(newTeam)
    ovr_player_data['team'].to_csv('!Data/Team.csv')
else:
    print("Columns do not match")
```
