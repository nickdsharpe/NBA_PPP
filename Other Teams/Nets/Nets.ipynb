```python
import pandas as pd
import time
import json
import numpy as np
import nbimporter
from PPP import PPP
np.seterr(invalid='ignore')
newGame = 'Nuggets3.12.23'
```

```python
try:
    newBridgesFile = (f'Bridges/Bridges{newGame}.csv')
    newBridges = pd.read_csv(newBridgesFile)
    newBridges = newBridges.dropna()
    newBridges = newBridges.set_index('Shot Type')
    newBridges = newBridges.astype('int')
except(FileNotFoundError):
    print('No Bridges File Found.')

bridgesFile = ('!Data/Bridges.csv')
bridges = pd.read_csv(bridgesFile)
bridges = bridges.dropna()
bridges = bridges.set_index('Shot Type')
bridges = bridges.astype('int')
```

```python
try:
    newCurryFile = (f'Curry/Curry{newGame}.csv')
    newCurry = pd.read_csv(newCurryFile)
    newCurry = newCurry.dropna()
    newCurry = newCurry.set_index('Shot Type')
    newCurry = newCurry.astype('int')
except(FileNotFoundError):
    print('No Curry File Found.')

curryFile = ('!Data/Curry.csv')
curry = pd.read_csv(curryFile)
curry = curry.dropna()
curry = curry.set_index('Shot Type')
curry = curry.astype('int')
```

```python
try:
    newDinwiddieFile = (f'Dinwiddie/Dinwiddie{newGame}.csv')
    newDinwiddie = pd.read_csv(newDinwiddieFile)
    newDinwiddie = newDinwiddie.dropna()
    newDinwiddie = newDinwiddie.set_index('Shot Type')
    newDinwiddie = newDinwiddie.astype('int')
except(FileNotFoundError):
    print('No Dinwiddie File Found.')

dinwiddieFile = ('!Data/Dinwiddie.csv')
dinwiddie = pd.read_csv(dinwiddieFile)
dinwiddie = dinwiddie.dropna()
dinwiddie = dinwiddie.set_index('Shot Type')
dinwiddie = dinwiddie.astype('int')
```

```python
try:
    newJohnsonFile = (f'Johnson/Johnson{newGame}.csv')
    newJohnson = pd.read_csv(newJohnsonFile)
    newJohnson = newJohnson.dropna()
    newJohnson = newJohnson.set_index('Shot Type')
    newJohnson = newJohnson.astype('int')
except(FileNotFoundError):
    print('No Johnson File Found.')

johnsonFile = ('!Data/Johnson.csv')
johnson = pd.read_csv(johnsonFile)
johnson = johnson.dropna()
johnson = johnson.set_index('Shot Type')
johnson = johnson.astype('int')
```

```python
try:
    newSimmonsFile = (f'Simmons/Simmons{newGame}.csv')
    newSimmons = pd.read_csv(newSimmonsFile)
    newSimmons = newSimmons.set_index('Shot Type')
except(FileNotFoundError):
    print('No Simmons File Found.')

simmonsFile = '!Data/Simmons.csv'
simmons = pd.read_csv(simmonsFile)
simmons = simmons.set_index("Shot Type")
```

```python
try:
    newThomasFile = (f'Thomas/Thomas{newGame}.csv')
    newThomas = pd.read_csv(newThomasFile)
    newThomas = newThomas.set_index('Shot Type')
except(FileNotFoundError):
    print('No Thomas File Found.')

thomasFile = '!Data/Thomas.csv'
thomas = pd.read_csv(thomasFile)
thomas = thomas.set_index("Shot Type")
```
