```python
import pandas as pd
import time
import json
import numpy as np
import nbimporter
from PPP import PPP
np.seterr(invalid='ignore')
newGame = 'Jazz3.7.23'
```

# Read in new files and load overall files

```python
try:
    newDoncicFile = (f'Doncic/Doncic{newGame}.csv')
    newDoncic = pd.read_csv(newDoncicFile)
    newDoncic = newDoncic.dropna()
    newDoncic = newDoncic.set_index('Shot Type')
    newDoncic = newDoncic.astype('int')
except(FileNotFoundError):
    print('No Doncic File Found.')

DoncicFile = ('!Data/Doncic.csv')
doncic = pd.read_csv(DoncicFile)
doncic = doncic.dropna()
doncic = doncic.set_index('Shot Type')
doncic = doncic.astype('int')
```

```python
try:
    newGreenFile = (f'Green/Green{newGame}.csv')
    newGreen = pd.read_csv(newGreenFile)
    newGreen = newGreen.dropna()
    newGreen = newGreen.set_index('Shot Type')
    newGreen = newGreen.astype('int')
except(FileNotFoundError):
    print('No Green File Found.')

greenFile = ('!Data/Green.csv')
green = pd.read_csv(greenFile)
green = green.dropna()
green = green.set_index('Shot Type')
green = green.astype('int')
```

```python
try:
    newHardawayFile = (f'Hardaway/Hardaway{newGame}.csv')
    newHardaway = pd.read_csv(newHardawayFile)
    newHardaway = newHardaway.dropna()
    newHardaway = newHardaway.set_index('Shot Type')
    newHardaway = newHardaway.astype('int')
except(FileNotFoundError):
    print('No Hardaway File Found.')

hardawayFile = ('!Data/Hardaway.csv')
hardaway = pd.read_csv(hardawayFile)
hardaway = hardaway.dropna()
hardaway = hardaway.set_index('Shot Type')
hardaway = hardaway.astype('int')
```

```python
try:
    newIrvingFile = (f'Irving/Irving{newGame}.csv')
    newIrving = pd.read_csv(newIrvingFile)
    newIrving = newIrving.dropna()
    newIrving = newIrving.set_index('Shot Type')
    newIrving = newIrving.astype('int')
except(FileNotFoundError):
    print('No Irving File Found.')

IrvingFile = ('!Data/Irving.csv')
irving = pd.read_csv(IrvingFile)
irving = irving.dropna()
irving = irving.set_index('Shot Type')
irving = irving.astype('int')
```

# Add new game data to overall data

```python
if doncic.columns.equals(newDoncic.columns):
    doncic = doncic.add(newDoncic)
    doncic.to_csv('!Data/Doncic.csv')
else:
    print("Columns do not match")
```

```python
if irving.columns.equals(newIrving.columns):
    irving = irving.add(newIrving)
    irving.to_csv('!Data/Irving.csv')
else:
    print("Columns do not match")
```

```python
if green.columns.equals(newGreen.columns):
    green = green.add(newGreen)
    green.to_csv('!Data/Green.csv')
else:
    print("Columns do not match")
```

```python
if hardaway.columns.equals(newHardaway.columns):
    hardaway = hardaway.add(newHardaway)
    hardaway.to_csv('!Data/Hardaway.csv')
else:
    print("Columns do not match")
```
