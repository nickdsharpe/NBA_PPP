```python
import pandas as pd
import time
import json
import numpy as np
np.seterr(invalid='ignore')
import nbimporter
from PPP import PPP

newGame = 'Game6'
prevGames = ['Game1', 'Game2', 'Game3', 'Game4', 'Game5', 'Game6']
Opp = 'Nuggets'
Team = 'Suns'
```

```python
players = ['ayton', 'biyombo', 'booker', 'craig', 'durant', 'landale', 'lee', 'okogie', 'paul', 'payne', 'ross', 'shamet', 'warren', 'team', 'opp']
```

```python
def load_data(players):
    
    new_player_data = {}
    ovr_player_data = {}
    
    for game in prevGames: # Loop over all games in series
        
        for player in players: # Loop over all players on roster
            
            if player == 'team':

                try:
                    newplayerFile = (f'Team/{game + Opp}Team.csv')
                    newplayer = pd.read_csv(newplayerFile)
                    newplayer = newplayer.dropna()
                    newplayer = newplayer.set_index('Shot Type')
                    newplayer = newplayer.astype('int')
                    new_player_data[player] = newplayer
                except(FileNotFoundError):
                    newplayerFile = ('Team/empty.csv')
                    newplayer = pd.read_csv(newplayerFile)
                    newplayer = newplayer.set_index('Shot Type')

                    for df in new_player_data.values():

                        newplayer = newplayer.add(df)

                    new_player_data['team'] = newplayer
                    newplayer.to_csv(f'../{Opp}/Opp/{newGame[0:5] + Team}Opp.csv')
                    newplayer.to_csv(f'Team/{newGame}Team.csv')

                if game[0:5] == 'Game1':
                    ovr_player_data['team'] = newplayer
                    newplayer.to_csv(f'!Data/Team.csv')
                    continue
                    
                # Load overall data
                ovrplayerFile = (f'!Data/Team.csv')
                ovrplayer = pd.read_csv(ovrplayerFile)
                ovrplayer = ovrplayer.dropna()
                ovrplayer = ovrplayer.set_index('Shot Type')
                ovrplayer = ovrplayer.astype('int')
                ovr_player_data[player] = ovrplayer
                    
                if newplayer.columns.equals(ovrplayer.columns):
                    ovrplayer = ovrplayer.add(newplayer)
                    ovr_player_data[player] = ovrplayer
                    ovrplayer.to_csv('!Data/Team.csv')
                else:
                    print(f"{player.title()}'s columns do not match")
                continue

            if player == 'opp':

                try:
                    newplayerFile = (f'Opp/{game + Opp}Opp.csv')
                    newplayer = pd.read_csv(newplayerFile)
                    newplayer = newplayer.dropna()
                    newplayer = newplayer.set_index('Shot Type')
                    newplayer = newplayer.astype('int')
                    new_player_data[player] = newplayer

                except(FileNotFoundError):
                    print(f'No {player.title()} file found.')
                    
                if game[0:5] == 'Game1':
                    ovr_player_data['opp'] = newplayer
                    newplayer.to_csv(f'!Data/Opp.csv')
                    continue
                    
                # Load overall data
                ovrplayerFile = (f'!Data/Opp.csv')
                ovrplayer = pd.read_csv(ovrplayerFile)
                ovrplayer = ovrplayer.dropna()
                ovrplayer = ovrplayer.set_index('Shot Type')
                ovrplayer = ovrplayer.astype('int')
                ovr_player_data[player] = ovrplayer

                if newplayer.columns.equals(ovrplayer.columns):
                    ovrplayer = ovrplayer.add(newplayer)
                    ovr_player_data[player] = ovrplayer
                    ovrplayer.to_csv('!Data/Opp.csv')
                else:
                    print(f"{player.title()}'s columns do not match")
                continue
                
            ### Loading/Adding the player files ###
            try:
                newplayerFile = (f'{player.title()}/{game + Opp + player.title()}.csv')
                newplayer = pd.read_csv(newplayerFile)
                newplayer = newplayer.dropna()
                newplayer = newplayer.set_index('Shot Type')
                newplayer = newplayer.astype('int')
                new_player_data[player] = newplayer
            except(FileNotFoundError):
                print(f'No {player.title()} file found for {game}.')
                continue
          
            if game == 'Game1':
                ovr_player_data[player] = newplayer
                newplayer.to_csv(f'!Data/{player.title()}.csv')
                continue
                
            # Load overall data
            try:
                ovrplayerFile = (f'!Data/{player.title()}.csv')
                ovrplayer = pd.read_csv(ovrplayerFile)
                ovrplayer = ovrplayer.dropna()
                ovrplayer = ovrplayer.set_index('Shot Type')
                ovrplayer = ovrplayer.astype('int')
                ovr_player_data[player] = ovrplayer
            except(FileNotFoundError):
                ovrplayer = newplayer
                ovr_player_data[player] = ovrplayer
                ovrplayer.to_csv(f'!Data/{player.title()}.csv')
                continue
            
            if newplayer.columns.equals(ovrplayer.columns):
                ovrplayer = ovrplayer.add(newplayer)
                ovr_player_data[player] = ovrplayer
                ovrplayer.to_csv(f'!Data/{player.title()}.csv')
            else:
                print(f"{player.title()}'s columns do not match")  
            continue
            
    return new_player_data, ovr_player_data   

new_player_data, ovr_player_data  = load_data(players)
```
