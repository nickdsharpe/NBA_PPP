```python
import pandas as pd
import time
import json
import numpy as np
import nbimporter
from PPP import PPP
np.seterr(invalid='ignore')
newGame = 'Celtics3.1.23'
```

# Read in new game files and load overall files

```python
try:
    newGarlandFile = (f'Garland/Garland{newGame}.csv')
    newGarland = pd.read_csv(newGarlandFile)
    newGarland = newGarland.dropna()
    newGarland = newGarland.set_index('Shot Type')
    newGarland = newGarland.astype('int')
except(FileNotFoundError):
    print('No Garland File Found.')

garlandFile = ('!Data/Garland.csv')
garland = pd.read_csv(garlandFile)
garland = garland.dropna()
garland = garland.set_index('Shot Type')
garland = garland.astype('int')
```

```python
try:
    newLeVertFile = (f'LeVert/LeVert{newGame}.csv')
    newLeVert = pd.read_csv(newLeVertFile)
    newLeVert = newLeVert.dropna()
    newLeVert = newLeVert.set_index('Shot Type')
    newLeVert = newLeVert.astype('int')
except(FileNotFoundError):
    print('No LeVert File Found.')

levertfile = ('!Data/LeVert.csv')
levert = pd.read_csv(levertfile)
levert = levert.dropna()
levert = levert.set_index('Shot Type')
levert = levert.astype('int')
```

```python
try:
    newMitchellFile = (f'Mitchell/Mitchell{newGame}.csv')
    newMitchell = pd.read_csv(newMitchellFile)
    newMitchell = newMitchell.dropna()
    newMitchell = newMitchell.set_index('Shot Type')
    newMitchell = newMitchell.astype('int')
except(FileNotFoundError):
    print('No Mitchell File Found.')

mitchellfile = ('!Data/Mitchell.csv')
mitchell = pd.read_csv(mitchellfile)
mitchell = mitchell.dropna()
mitchell = mitchell.set_index('Shot Type')
mitchell = mitchell.astype('int')
```

```python
try:
    newMobleyFile = (f'Mobley/Mobley{newGame}.csv')
    newMobley = pd.read_csv(newMobleyFile)
    newMobley = newMobley.dropna()
    newMobley = newMobley.set_index('Shot Type')
    newMobley = newMobley.astype('int')
except(FileNotFoundError):
    print('No Mobley File Found.')

mobleyfile = ('!Data/Mobley.csv')
mobley = pd.read_csv(mobleyfile)
mobley = mobley.dropna()
mobley = mobley.set_index('Shot Type')
mobley = mobley.astype('int')
```

# Add new game data to overall files

```python
if garland.columns.equals(newGarland.columns):
    garland = garland.add(newGarland)
    garland.to_csv('!Data/Garland.csv')
else:
    print("Columns do not match")
```

```python
if levert.columns.equals(newLeVert.columns):
    levert = levert.add(newLeVert)
    levert.to_csv('!Data/LeVert.csv')
else:
    print("Columns do not match")
```

```python
if mitchell.columns.equals(newMitchell.columns):
    mitchell = mitchell.add(newMitchell)
    mitchell.to_csv('!Data/Mitchell.csv')
else:
    print("Columns do not match")
```

```python
if mobley.columns.equals(newMobley.columns):
    mobley = mobley.add(newMobley)
    mobley.to_csv('!Data/Mobley.csv')
else:
    print("Columns do not match")
```
